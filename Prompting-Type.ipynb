```python
from openai import OpenAI

Client = OpenAI(api_key="secrete key")
```

```python
def call_llm(messages, model = "gpt-4o-mini", temperature = 1):
  response = Client.chat.completions.create(
    model = model,
    messages = messages,
    temperature = temperature
  )
  return response.choices[0].message.content
```

```python
#zero-shot promting
messages = [
    {
        "role": "user",
        "content": "Explain cyber security."
    }
]

print(call_llm(messages))
```

```python
#one-shot prompting
messages = [
    {
        "role": "user",
        "content": """Q: What is a stack?
        A: A stack is a linear data structure that follows LIFO.

        Q: What is a queue?
        A:"""

    }
]

print(call_llm(messages))
```

```python
#Few-shot prompting
messages = [
    {
        "role": "user",
        "content": """English: Apple
French: Pomme

English: Book
French: Livre

English: Computer
French:"""

    }
]

print(call_llm(messages))
```

```python
#Instruction-Based Prompting
messages = [
    {
        "role": "user",
        "content": "Summarize the importance of cyber security in exactly 3 lines."

    }
]

print(call_llm(messages))
```

```python
#Role-Based Prompting
messages = [
    {
        "role": "user",
        "content": """You are a cyber security expert.
Explain phishing attacks with a real-life example."""

    }
]

print(call_llm(messages))
```
