```python
from openai import OpenAI

client = OpenAI(api_key=api_key)

def call_gpt4o_mini(prompt, temperature = 0.2):
  response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are a helpful AI assistant." },
        {"role": "user", "content": prompt },
    ],
    temperature=temperature,
  )
  return response.choices[0].message.content
```

```python
def chain_of_thougth(question):
  prompt = f"""
  Solve the problem step by step and explain your reasoning.
  Question: {question}

  Let's think step by step
  """

  return call_gpt4o_mini(prompt)
```

```python
print(chain_of_thougth(
    "A cyclist rides 90 km in 3 hours .What is the average speed?"
))
```

```python
def few_shot_sentiment(text):
  prompt = f"""
  Classify the sentiment as Positive, Negative, or Neutral.

  Example 1:
  Text: I love this movie.
  Sentiment: Positive

  Example 2:
  Text: This is the worst movie.
  Sentiment: Negative

  Example 3:
  Text: The delivery arrived on time.
  Sentiment: Neutral

  Now classify:
  Text: {text}
  Sentiment:
  """
  return  call_gpt4o_mini(prompt,temperature = 0)
```

```python
print(few_shot_sentiment("I hate this movie."))
```
